```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
data_csv = pd.read_csv('https://raw.githubusercontent.com/pikimen357/Proyek-Statisika/refs/heads/main/toughestsport.csv')
data_csv
```

```python
# Distribution of Total scores
plt.figure(figsize=(6, 3))
plt.subplot(1, 2, 1)
sns.histplot(data_csv['Total'], bins=15, kde=True)
plt.title('Distribution of Total Scores')

plt.tight_layout()
plt.show()
```

```python
skills = ['Endurance', 'Strength', 'Power', 'Speed', 'Agility', 'Flexibility', 'Nerve', 'Durability', 'Hand-eye coordination', 'Analytical Aptitude']
plt.figure(figsize=(5, 5))
sns.boxplot(data_csv[skills], orient='h')
plt.xlabel('Skill Score')
plt.ylabel('Skill')
plt.title('Comparison of Skill Scores')
plt.show()
```

```python
# Correlation matrix
corr = data_csv[skills].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Skills')
plt.show()
```

```python
# Assuming 'data' is your DataFrame containing the dataset
# 'skills' is the list of skills and 'Rank' is the column containing the rank of each sport
correlations = data_csv[skills + ['Rank']].corr()['Rank'].drop('Rank')

# Print the correlations
print("Correlation of skills with Rank:")
print(correlations)
```

```python
fig, axes = plt.subplots()

sns.barplot(x=data_csv.iloc[:5, 0], y=data_csv.iloc[:5, -2], ax=axes, palette='inferno') # Ganti 'Purl' dengan nama palet lain jika diinginkan ('Blues', 'Greens', 'Reds', 'Purples', 'Oranges', 'viridis', 'plasma', 'inferno', 'magma', 'cividis'.)

for container in axes.containers:
    axes.bar_label(container, label_type="center")
plt.show()
```

```python
# Observing the correlation between Endurance and Rank
# X is selected as a DataFrame with one column
X = data_csv[['Endurance']]
y = data_csv['Rank']

# 3. Model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# 4. Evaluasi
print("Koefisien regresi:")
# Iterate through the columns of the X DataFrame for printing
for feature, coef in zip(X.columns, model.coef_):
    print(f"{feature}: {coef:.4f}")

strength_ranking_correlation = data_csv['Endurance'].corr(data_csv['Rank'])

print(f"Correlation between Endurance and Rankings: {strength_ranking_correlation}")
print(f"\nIntercept: {model.intercept_:.4f}")
print(f"R score: {r2_score(y, y_pred):.4f}")
print(f"Mean Squared Error: {mean_squared_error(y, y_pred):.4f}")

# 5. Scatter Plot: Actual vs Predicted
plt.figure(figsize=(8,6))
plt.scatter(y, y_pred, color='royalblue', edgecolor='black')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', linewidth=2)
plt.xlabel('Endurance')
plt.ylabel('Rank')
plt.title(' Endurance VS Rank (Linear Regression)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
